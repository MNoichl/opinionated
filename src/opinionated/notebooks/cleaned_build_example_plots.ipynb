## Basic Scatter Plot

```python
# Set the style for the plots
plt.style.use('default')

# Load the penguins dataset from seaborn
penguins = sns.load_dataset("penguins")

# Create a scatter plot of bill length against flipper length
f, ax = plt.subplots(figsize=(10, 7))
# plt.scatter(x=penguins["bill_length_mm"], y=penguins["flipper_length_mm"], c=pd.factorize(penguins["species"])[0])
sns.scatterplot(x="bill_length_mm", y="flipper_length_mm", hue="species", data=penguins, alpha=.7, s= 70)

# Set the title for the plot
ax.set_title('Penguins')
plt.savefig('img/outofthebox_penguins.png', bbox_inches='tight', pad_inches=.35, dpi=400, transparent=False)
```

## Plot example plots for different styles

```python
# Set the style for the plots
import opinionated
import colormaps as cmaps

style_names = [
    "opinionated_rc",
    "opinionated_fsc",
    "opinionated_j",
    "opinionated_m",
    "opinionated_sg",
    "opinionated_tw",
]

for this_style in style_names:

    plt.style.use(this_style)
    # Create a scatter plot of bill length against flipper length
    f, ax = plt.subplots(figsize=(10, 7))
    # plt.scatter(x=penguins["bill_length_mm"], y=penguins["flipper_length_mm"], c=pd.factorize(penguins["species"])[0],cmap=cmaps.colorblind_10[0:3], alpha=.7)

    sns.scatterplot(x="bill_length_mm", y="flipper_length_mm", hue="species", data=penguins, alpha=.7, s= 70, palette=cmaps.bold[2:5]._colors)


    opinionated.add_legend(title='Species')
    opinionated.add_attribution('by Maximilian Noichl')
    opinionated.set_title_and_suptitle('Penguins!','They are an excellent type of bird!')
    plt.savefig('img/'+ this_style + '_example.png',bbox_inches='tight', pad_inches=.35, dpi =400, transparent=False)
```

```python
  plt.style.use("opinionated_fsc")

flights = sns.load_dataset("flights")
g = sns.relplot(
    data=flights,
    x="month", y="passengers", col="year", hue="year",
    kind="line", palette=cmaps.matter , linewidth=4, zorder=5,
    col_wrap=3, height=2, aspect=1.5, legend=False,
)

for year, ax in g.axes_dict.items():
    ax.text(.8, .85, year, transform=ax.transAxes, fontweight="bold")
    sns.lineplot(
        data=flights, x="month", y="passengers", units="year",
        estimator=None, color=".7", linewidth=1, ax=ax,
    )

ax.set_xticks(ax.get_xticks()[::2])
g.set_titles("")

g.set_axis_labels("", "")
g.tight_layout()

opinionated.add_attribution('by Maximilian Noichl', position = [.97, -0.01])
# opinionated.set_title_and_suptitle('Flights!','More of them with every year!',1.1)
plt.figtext( .06, .97 * 1.1, 'Flights!', fontsize=26, fontweight='bold', ha='left')
plt.figtext( .1, .918 * 1.1, 'More of them with every year!',  fontsize=14, fontweight='regular', ha='left')

plt.savefig('img/opinions_fira_sans_condensed_example.png',bbox_inches='tight',
            dpi =400, transparent=False)
```
