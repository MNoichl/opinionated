```python
import seaborn as sns
from matplotlib import font_manager as fm, pyplot as plt

import shutil
import matplotlib
import pandas as pd


plt.style.use('default')
#Loading the data:

penguins = sns.load_dataset("penguins")


f, ax = plt.subplots(figsize=(10, 7))
sns.scatterplot(x="bill_length_mm", y="flipper_length_mm", data=penguins,
            style='species',c='black')

ax.set_title('Penguins')

plt.savefig('gdrive/MyDrive/opinionated/img/outofthebox_penguins.png',bbox_inches='tight',dpi =400, transparent=False)
```
